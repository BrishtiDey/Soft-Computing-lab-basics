# Bell shaped curve (23 jan 2023)

```python
import numpy as np
import matplotlib.pyplot as plt


def bell(x):
    mean = np.mean(x)
    std = np.std(x)
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp(- (x - mean)*2 / (2 * std*2))
    return y_out


x = np.arange(-2, 2, 0.1)
y = bell(x)
plt.plot(x, y)
plt.show()
```

# Trapezoidal

```python
# Trapezoidal
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-30, 100, 500)


def trap(points, x):
    values = []
    for x1 in x:
        if x1 <= points['a']:
            values.append(0)
        elif (x1 >= points['a'] and x1 <= points['b']):
            values.append((x1-points['a'])/(points['b'] - points['a']))
        elif x1 >= points['b'] and x1 <= points['c']:
            values.append(1)
        elif (x1 >= points['c'] and x1 <= points['d']):
            values.append((points['d']-x1)/(points['d'] - points['c']))
        elif x1 >= points['d']:
            values.append(0)
    return values


points = {'a': -15,  'b': -9, 'c': 70,  'd': 80}
plt.plot(x, trap(points, x))
```

# Triangular

```python
# Triangular
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-50, 100, 500)


def tri(points, x):
    values = []
    for x1 in x:
        if x1 <= points['a']:
            values.append(0)
        elif (x1 >= points['a'] and x1 <= points['b']):
            values.append((x1-points['a'])/(points['b'] - points['a']))
        elif x1 >= points['b'] and x1 <= points['c']:
            values.append((points['c']-x1)/(points['c'] - points['b']))
        elif x1 >= points['c']:
            values.append(0)
    return values


tripoints = {'a': -15,  'b': 20,   'c': 50}
plt.plot(x, tri(tripoints, x))
```

```python
# Two points Crossover 13/02/2023
import random
A = [1, 0, 0, 1, 1, 0, 1, 0]  # Inpit 1
B = [0, 0, 1, 1, 0, 0, 0, 1]  # Input 2
print(A, B)

cp1 = random.randint(1, len(A)-2)
cp2 = random.randint(cp1+1, len(A)-2)
print(cp1, cp2)

for i in range(cp1+1, cp2):
    A[i], B[i] = B[i], A[i]

print(A,B)
```
