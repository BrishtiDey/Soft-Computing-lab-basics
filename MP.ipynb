# Norml distribution

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)*2 / (2 * sigma*2) ),linewidth=3, color='y')
plt.show()
```

# Trapezoidal

```python

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-30, 100, 500)


def trap(points, x):
    values = []
    for x1 in x:
        if x1 <= points['a']:
            values.append(0)
        elif (x1 >= points['a'] and x1 <= points['b']):
            values.append((x1-points['a'])/(points['b'] - points['a']))
        elif x1 >= points['b'] and x1 <= points['c']:
            values.append(1)
        elif (x1 >= points['c'] and x1 <= points['d']):
            values.append((points['d']-x1)/(points['d'] - points['c']))
        elif x1 >= points['d']:
            values.append(0)
    return values


points = {'a': -15,  'b': -9, 'c': 70,  'd': 80}
plt.plot(x, trap(points, x))
```

# Triangular

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-50, 100, 500)


def tri(points, x):
    values = []
    for x1 in x:
        if x1 <= points['a']:
            values.append(0)
        elif (x1 >= points['a'] and x1 <= points['b']):
            values.append((x1-points['a'])/(points['b'] - points['a']))
        elif x1 >= points['b'] and x1 <= points['c']:
            values.append((points['c']-x1)/(points['c'] - points['b']))
        elif x1 >= points['c']:
            values.append(0)
    return values


tripoints = {'a': -15,  'b': 20,   'c': 50}
plt.plot(x, tri(tripoints, x))
```

# Min composition

```python
import matplotlib.pyplot as plt
import numpy as np

def gaus(x,x2,a):
    return(np.exp(-0.5*((x-x2)/a)**2))
def mf2d(x,y):
    return min(gaus(x,50,5),gaus(y,50,5))

f=np.vectorize(mf2d)
fig=plt.figure()
ax=plt.axes(projection='3d')
x=np.linspace(0,100,500)
y=np.linspace(0,100,500)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax.plot_surface(X,Y,Z)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
# New Antecedent/Consequent objects hold universe variables and membership
# functions
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')
# Auto-membership function population is possible with.automf(3, 5, or 7)
quality.automf(3)
service.automf(3)
# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
```

```python
quality['average'].view()
```

```python
service.view()
```

```python
tip.view()
```

```python
rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])
rule1.view()
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['quality'] = 6.5
tipping.input['service'] = 9.8
# Crunch the numbers
tipping.compute()
```

```python
print(tipping.output['tip'])
tip.view(sim=tipping)
```
